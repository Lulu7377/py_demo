```python
# if else statement

# cat 1 >90
# cat 2 >70
# cat 3 <70

# using list comprehension

final_tbl['category'] = ['cat a' if a >= 90 else 'cat 2' if a < 90 and a >= 70 else 'cat 3' for a in final_tbl.MARKS ]

# second way forward is 
final_tbl['category_1'] = 'cat 3'
final_tbl['category_1'][final_tbl.MARKS >= 90] = 'cat a'
final_tbl['category_1'][(final_tbl.MARKS <90)&(final_tbl.MARKS >= 70)] = 'cat 2'

# third way forward is 
import numpy as np

final_tbl['category_2'] = np.where(final_tbl.MARKS >= 90, 'cat a', np.where(final_tbl.MARKS >= 70, 'cat 2', 'cat 3'))
```
